```python
import random

def play_hangman():
    words = ["Bamboozle","Shenanigans","Skedaddle","Kerfuffle","Hullabaloo"]
    target_word = random.choice(words)
    
    display = ["_"] * len(target_word)
    guessed_letters = []
    attempts_remaining = 6

 
    print("You have 6 incorrect guesses allowed.\n")

    while attempts_remaining > 0 and "_" in display:
        print("Word:", " ".join(display))
        print("\033[33mAttempts remaining:", attempts_remaining, "\033[0m")
        print("Guessed letters:", ", ".join(guessed_letters))
        
        guess = input("Guess a letter: ").lower()

        if len(guess) != 1 or not guess.isalpha():
            print("❌ Please enter a single valid letter.\n")
            continue

        if guess in guessed_letters:
            print("⚠️ You already guessed that letter.\n")
            continue

        guessed_letters.append(guess)

        if guess in target_word:
            print("\033[32m✅ Correct guess!\033[0m")
            for i in range(len(target_word)):
                if target_word[i] == guess:
                    display[i] = guess
        else:
            attempts_remaining -= 1
            print("\033[31m❌ Wrong guess!\033[0m")

    if "_" not in display:
        print("🎉 Congratulations! You guessed the word:", target_word)
    else:
        print("\033[35m💀 Game Over!\033[0m", target_word)

if __name__ == "__main__":
    play_hangman()
```
